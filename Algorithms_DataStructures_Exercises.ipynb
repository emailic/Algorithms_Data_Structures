`````python
## Your implementation here!!!
def SelectionSort(coll):
    A = list(coll)
    for i in range(len(A)):
        min_idx = i
        for j in range(i+1, len(A)):
            if A[min_idx] > A[j]:
                min_idx = j
                
        # Swap 
        A[i], A[min_idx] = A[min_idx], A[i]
    return A
`````

You have to implements Insertion Sort algorithm.

`````python
## Your implementation here!!!

def InsertionSort(coll):
    
    A = list(coll)
    for i in range(len(A)):
        c = A[i] #current element
        j = i-1
        #print('outer loop',curr,j,i)
        while j >= 0 and c < A[j]:
            A[j+1] = A[j]
            j -= 1
            #print('inner loop',curr,j,i)
        A[j+1] = c #this is how we modify the list instead of creating a new one
        
    return A
`````

Write the function ```InsertionSort(coll, cmp)``` that returns a sorted list with the elements in *coll* using 
```cmp```as a comparator.

`````python
## Your implementation here!!!

def InsertionSort(coll, cmp):
    A = list(coll)
    for i in range(1, len(A)):
        c = A[i]
        j = i
        while j > 0 and cmp(A[j-1], c)>0:
            A[j] = A[j-1]
            j -= 1
        A[j] = c
    return A  
`````

Below an implementation of QuickSort. 

In this exercise you have to:

`````python
## Your implementation here!!!
def partition(A, low, high): #takes the array, the beginning and the end of the subarray
    pivot = A[high]
    #we are selecting the element at the ending index to be our pivot point
    i = low-1 
    #index of a smaller element
  
    for j in range(low, high): 
        #iterating indices between the beginning and the end of the list:
        if A[j] <= pivot: 
            #if the given element is smaller than or equal to the pivot point element:
            i = i+1 
            #we increase the index of the smaller element by one
            A[i], A[j] = A[j], A[i] 
            #and we swap the smaller element with the given element.
  
    A[i+1], A[high] = A[high], A[i+1] 
    return i+1

def randPartition(A,low,high):
    pi = low # pivot
    
    # we use i to know where the partition in the array starts from
    i = low + 1
    

    for j in range(low + 1, high + 1):
        
        # if the  element in question is smaller or equal to pivot, we swap it with the right element.
        if A[j] <= A[pi]:
            A[i] , A[j] = A[j] , A[i]
            i = i + 1
    A[pi] , A[i - 1] =A[i - 1] , A[pi]
    pi = i - 1
    return (pi)

def quickSort_rec(A, low, high):
    if low < high: 
        #pi is a partitioning index and thus the A[pi] is at a right place.
        pi = randPartition(A, low, high) 
        quickSort_rec(A, low, pi-1) 
        quickSort_rec(A, pi+1, high) 
        
def quickSort(B): #this function helps to avoid the call to the previous funciton where
                  #we have to remmeber the beginning and the end.
    A = B[:] # Copy the array just because we decided to return a sorted copy of the original array 
    quickSort_rec(A, 0, len(A)-1)
    return A
`````

`````python
## Your implementation here!!!
def merge(myList, left, right):
    #iterators for iterating the halves
    i = 0
    j = 0

    #main interator
    k = 0
    #print('left and right: ', left, right)
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            myList[k] = left[i]
            i += 1
        else:
            myList[k] = right[j]
            j += 1
        k += 1

    while i < len(left):
        myList[k] = left[i]
        i += 1
        k += 1

    while j < len(right):
        myList[k]=right[j]
        j += 1
        k += 1
    #print('end of merge', myList)
    return myList

def mergeSort_rec(myList):
    if len(myList) > 1:
        mid = len(myList) // 2
        left = myList[:mid]
        right = myList[mid:]
        #print(myList, left, right)
        # Recursive call on each half
        mergeSort_rec(left)
        mergeSort_rec(right)
        merge(myList, left, right)
        
def mergeSort(B):
    A = B[:] # Copy the array just because we decided to return a sorted copy of the original array 
    mergeSort_rec(B)
    #print('the array',A)
    return (B)
`````

`````python
## Your implementation here!!!
## Your implementation goes here
#(v,w)
def fractional_knapsack(L,W):
    def ratio(L):
        ratio=dict()
        for elem in L:
            ratio[elem]=( elem[0]/elem[1])
        return ratio
    r=sorted(ratio(L).items(),key=lambda x: x[1], reverse=True)
    sumw=0  #sum of weight
    sumv=0 #sum of value
    for i in range (len(r)):
        if sumw<=W:
            sumw+=list(r)[i][0][1] #add weight
            sumv+=list(r)[i][0][0] #add value
            memory=i
    sumv-=list(r)[memory][0][0] #remove the value of the last elemnt we added
    sumw-=list(r)[memory][0][1] #remove the weight of the last elemnt we added
    diff=(W-sumw)/list(r)[memory][0][1]
    adjv=diff*(list(r)[memory][0][0]) #calculate the proportion of the value to add
    return sumv+adjv
    
    

            

`````

A point $(x,y)$ dominates a point $(x',y')$ iff $𝑥'\leq 𝑥$ and $y'\leq 𝑦$.

This step can be performed in linear time.

### Exercise: Strongly Connected Components
The goal of this exercise is to implement the following algorithm to compute the strongly connected components of a directed graph $G$. [See here](https://www.hackerearth.com/practice/algorithms/graphs/strongly-connected-components/tutorial/). 

In a directed graph a component is strongly connected if there is a directed path from any vertex to every other vertex of the component. The problem asks to is to partition the graph into maximal strongly connected components.

NetworkX provides a method to compute the strongly connected components of a graph [here](https://networkx.github.io/documentation/stable/reference/algorithms/generated/networkx.algorithms.components.strongly_connected_components.html#networkx.algorithms.components.strongly_connected_components). 

Your goal is to implement the Kosaraju's Linear time algorithm to find Strongly Connected Component. 
The algorithm is described [here](https://www.hackerearth.com/practice/algorithms/graphs/strongly-connected-components/tutorial).

It works in three steps. 

- Do a DFS on the original graph, keeping track of the DFS finish times of each node. This can be done with a stack, when some  finishes put the source vertex on the stack. This way node with highest finishing time will be on top of the stack.
- Reverse the original graph, i.e., if there is an edge $(u,v)$ in the original graph, add the edge $(v,u)$ in the reversed one.
-  Do DFS on the reversed graph, with the source vertex as the vertex on top of the stack. When DFS finishes, all nodes visited will form one Strongly Connected Component. If any more nodes remain unvisited, this means there are more Strongly Connected Component's, so pop vertices from top of the stack until a valid unvisited node is found. This will have the highest finishing time of all currently unvisited nodes.

Take a look at [DFS traversal documentation](https://networkx.github.io/documentation/stable/reference/algorithms/traversal.html#module-networkx.algorithms.traversal.depth_first_search).
Note that the finishing time of a node can be inferred from its position in the DFS tree. 

`````python
## Your implementation here!!!

class Graph:
    def __init__(self,vertices):
        self.V= vertices #No. of vertices
        self.graph=dict()


    # fill out a graph
    def addEdge(self,u,v):
        if u in self.graph.keys():
            self.graph[u]+=[v]
        else:
            self.graph[u]=[v]
        if v not in  self.graph.keys():
            self.graph[v]=[]
                
        #self.graph[u].append(v)

    # STEP1: DFS ON THE ORIGINAL GRAPH
    def DFS(self,v,visited):
        # Mark the current node as visited and print it
        visited[v]= True
        print ( v, end=', ')
        #Recur for all the vertices adjacent to this vertex
        #print(v, self.graph[v], self.graph)
        for i in self.graph[v]:
            #print(v, self.graph[v])
            if visited[i]==False:
                self.DFS(i,visited)

    #mark visited nodes as such to avoid infinite loops 
    def visitedN(self,v,visited, stack):
        visited[v]= True #current node is visited
        #Recur for all the vertices adjacent to this vertex
        #print(v, self.graph[v], self.graph)
        for i in self.graph[v]:
            if visited[i]==False:
                self.visitedN(i, visited, stack)
        stack = stack.append(v)
        #print('stack:',stack)


    # STEP 2: REVERSE THE GRAPH
    def reverse(self):
        g = Graph(self.V)

        # Recur for all the vertices adjacent to this vertex
        for i in self.graph:
            for j in self.graph[i]:
                g.addEdge(j,i)
        return g



    # STEP 3: find all the strongly connected components
    def printSCCs(self):

        stack = [] #all the vertices

        visited =[False]*(self.V) #mark all vertices as unvisited
        #add to the stack  vertices according to their finish times
        for i in range(self.V):
            if visited[i]==False:
                self.visitedN(i, visited, stack)

        
        gr = self.reverse()

        # again mark everything as unvisited for the second dfs
        visited =[False]*(self.V)


        cn=1 #component number
        #print('stack:',stack)
        while stack:
            i = stack.pop() #assign the value of the vertice to i and remove it from stack
            if visited[i]==False:
                print('Component number',cn,': ' , end='')
                gr.DFS(i, visited)
                print('')
                cn+=1
`````
